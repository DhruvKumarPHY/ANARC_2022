```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.rcParams['figure.figsize'] = [12, 8]
plt.rcParams.update({'font.size': 18})
```

```python
# spectral graph between time and 
dt = 0.0001
t = np.arange(0,3,dt)
t1 = 3
```

```python
t # Numpy array.
```

```python
np.power(t,2) # whole (t) array is squared wsing np.power 
```

```python
# Frequency 
f0 = 100
f1 = 500
```

```python
x = np.cos(2*np.pi*t*(f0 + (f1-f0)*np.power(t,2)/(3*t1**2)))
```

```python
x
```

```python
plt.plot(x)
```

```python
plt.specgram(x, NFFT=128, Fs=1/dt, noverlap=120,cmap='jet')
plt.colorbar()
plt.xlabel("$Time (s)$")
plt.ylabel("$Frequency (Hz)$")
plt.show()
```

```python
# magnified view....
plt.specgram(x, NFFT=128, Fs=1/dt, noverlap=120,cmap='jet')
plt.ylim(0,2000)
plt.xlabel("$Time (s)$")
plt.ylabel("$Frequency (Hz)$")
plt.colorbar()
plt.show()
```
