```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define the function
dx = 0.0001
L = 2*np.pi
x = np.arange(0,L+dx,dx)
n = len(x)
n_quad = int(np.floor(n/4)) # Greatest Integer 
```

```python
x
```

```python
f = np.zeros_like(x) # same dimensions as x
f[n_quad:3*n_quad] = 1
```

```python
plt.plot(x,f) # This is our desired function 
```

```python
# Let's bring out some Fourier
Ao = (np.sum(f* np.ones_like(x)) * dx * 2)/L
ffs = (Ao/2) * np.ones_like(f)
```

```python
for k in range(1,501):
    Ak = np.sum(f * np.cos(2*np.pi*k*x/L)) * dx * 2 / L
    Bk = np.sum(f * np.sin(2*np.pi*k*x/L)) * dx * 2 / L
    ffs = ffs + Ak*np.cos(2*k*np.pi*x/L) + Bk*np.sin(2*k*np.pi*x/L)
```

```python
plt.plot(x,f,color='k',linewidth=2) # discontuinity in the function is shown by these spiked lines.
plt.plot(x,ffs,'-',color='r',linewidth=1.5)
plt.show()
```
