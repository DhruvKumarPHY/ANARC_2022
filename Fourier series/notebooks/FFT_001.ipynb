# Parameters for the Function

```python
# define the function f(x)
dx = 0.00001
L = np.pi # period 
x = L * np.arange(-1+dx,1+dx,dx) # between (-1+dx,1+dx) take equally placed interval with dx as its interval.
n = len(x)
n_quad = int(np.floor(n/4))
```

```python
f = np.zeros_like(x)
```

```python
f[n_quad:2*n_quad] = (np.arange(1,n_quad+1))/n_quad    # here, (1--> 50,000 and n_quad+1 ---> 100,000)
# 50,000 elements in np.arange() and for whole argument...
f[2*n_quad:3*n_quad] = np.ones(n_quad) - ((np.arange(0,n_quad))/n_quad)
```

# Compute Fourier series

```python
Ao = np.sum(f * np.ones_like(x)) * dx
ffs = Ao/2 
```

```python
f * np.ones_like(x) 
```

```python
np.ones_like(x) # np array of the same size as x
```

```python
# Fourier Coffecients
j = 10
A = np.zeros(j)
B = np.zeros(j)
for k in range(j):
    A[k] = np.sum(f* np.cos(np.pi*(k+1)*x/L)) * dx
    B[k] = np.sum(f* np.sin(np.pi*(k+1)*x/L)) * dx
    ffs = ffs + A[k]*np.cos((k+1)*np.pi*x/L) + B[k]*np.sin((k+1)*np.pi*x/L) 
```

```python
plt.plot(x,f,color='k') # actual graph
plt.plot(x,ffs,'--',color='red') # fft 
```
